```python
fcu, fcu_freq  = read_fcu("../output/frenchCropUsage_20210525/d2kab/fcu-baseline.csv", "../resources/Corpus-v-12-04-21/train/")
```

```python
# prepare headers annotations
fcu['header'] = 'false'
fcu.loc[fcu['html tag'].str.startswith('H', na=False),'header'] = 'true'
headers = fcu[['bsv', 'number of words','prefLabel', "header"]]
headers['bsv'] = headers['bsv'].str.split(r'\/').str.get(-1)
headers['bsv'] = headers['bsv'].str.split(r'\.').str.get(-2)
headers.tail()
```

```python
headers.loc[headers['header'] == 'true']
```

```python
# load bm25 scores
scores = pd.read_csv("../output/frenchCropUsage_20210525/d2kab/bm25-baseline.csv", sep = '\t', encoding = 'utf8', names = ['bsv', 'prefLabel', 'score'])
scores.tail()
```

```python
# add scores to fcu vocabulary
result = pd.merge(headers, scores, on=["bsv", "prefLabel"])
result
```

```python
pd.DataFrame({'count' : result.groupby( ['bsv','prefLabel', 'score','header'] ).size().unstack(fill_value=0).stack()}).reset_index()
```

```python
test = pd.DataFrame({'count' : result.groupby( ['bsv','prefLabel', 'score','header'] ).size().unstack(fill_value=0).stack()}).reset_index()
test['total'] = test.groupby(['bsv', 'prefLabel','score'])['count'].transform('sum')
test['percentage'] = test.groupby(['bsv', 'prefLabel','score'])['count'].transform(lambda x: x/x.sum() * 100)
test
```

```python
# MEAN
```

```python
# mean for headers
df1 = test[test.header == "true"].sort_values(by=['score', 'percentage'])
df1[["score", "percentage"]].mean()
```

```python
# mean for when labels are not in headers
df2 = test[test.header == "false"]
df2[["score", "percentage"]].mean()
```
